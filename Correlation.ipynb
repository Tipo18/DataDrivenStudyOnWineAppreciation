```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the Excel file (.xlsx)
excel_file_path = "WineRankingDataWIPTomTim.xlsx"
data = pd.read_excel(excel_file_path)

# Calculate the correlation between "Price" and "Grade"
correlation = data['Price'].corr(data['Grade'])

# Display the correlation
print(f"Correlation between 'Price' and 'Grade': {correlation}")

# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(data['Grade'], data['Price'], alpha=0.5, label='Data Points')
plt.title("Scatter Plot of Price vs Grade with Trendline")
plt.xlabel("Grade")
plt.ylabel("Price")

# Add a trendline (first-degree polynomial)
coefficients = np.polyfit(data['Grade'], data['Price'], 1)
trendline = np.poly1d(coefficients)
plt.plot(data['Grade'], trendline(data['Grade']), color='red', label='Trendline')

plt.legend()
plt.grid(True)
plt.show()

```
