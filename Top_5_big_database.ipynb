```python
import pandas as pd

# Specify the file path
excel_file_path = "winemag-data-130k-v2.xlsx"

# Read only the selected columns from the Excel file
selected_columns = ['winery', 'province', 'country', 'variety', 'points']
data_subset = pd.read_excel(excel_file_path, usecols=selected_columns)
```

```python

# Top 5 elements with the highest frequency for each column
top_winery_count = data_subset['winery'].value_counts().nlargest(5)
top_province_count = data_subset['province'].value_counts().nlargest(5)
top_country_count = data_subset['country'].value_counts().nlargest(5)
top_variety_count = data_subset['variety'].value_counts().nlargest(5)

# Calculate the average points for each of the top elements
top_winery_avg_points = data_subset.groupby('winery')['points'].mean().loc[top_winery_count.index]
top_province_avg_points = data_subset.groupby('province')['points'].mean().loc[top_province_count.index]
top_country_avg_points = data_subset.groupby('country')['points'].mean().loc[top_country_count.index]
top_variety_avg_points = data_subset.groupby('variety')['points'].mean().loc[top_variety_count.index]

# Display the results
print("Top 5 wineries, province, countries and varieties with the highest frequency and their average points:\n")
result_df_winery = pd.DataFrame({
    'Winery': top_winery_count.index,
    'Frequency': top_winery_count.values,
    'Average Points': top_winery_avg_points.values
})
print(result_df_winery)
print()

result_df_province = pd.DataFrame({
    'Province': top_province_count.index,
    'Frequency': top_province_count.values,
    'Average Points': top_province_avg_points.values
})
print(result_df_province)
print()

result_df_country = pd.DataFrame({
    'Country': top_country_count.index,
    'Frequency': top_country_count.values,
    'Average Points': top_country_avg_points.values
})
print(result_df_country)
print()

result_df_variety = pd.DataFrame({
    'Variety': top_variety_count.index,
    'Frequency': top_variety_count.values,
    'Average Points': top_variety_avg_points.values
})
print(result_df_variety)
```

```python


# Calculate the average points for each element
avg_points_by_winery = data_subset.groupby('winery')['points'].mean()
avg_points_by_province = data_subset.groupby('province')['points'].mean()
avg_points_by_country = data_subset.groupby('country')['points'].mean()
avg_points_by_variety = data_subset.groupby('variety')['points'].mean()

# Display the top 5 elements with the highest average points and their frequencies
print("Top 5 wineries, province, countries and varieties with the highest average points and their frequencies:\n")
top_winery_df = pd.DataFrame({
    'Average Points': avg_points_by_winery.nlargest(5),
    'Frequency': data_subset['winery'].value_counts().loc[avg_points_by_winery.nlargest(5).index]
})
print(top_winery_df)

top_province_df = pd.DataFrame({
    'Average Points': avg_points_by_province.nlargest(5),
    'Frequency': data_subset['province'].value_counts().loc[avg_points_by_province.nlargest(5).index]
})
print(top_province_df)

top_country_df = pd.DataFrame({
    'Average Points': avg_points_by_country.nlargest(5),
    'Frequency': data_subset['country'].value_counts().loc[avg_points_by_country.nlargest(5).index]
})
print(top_country_df)
top_variety_df = pd.DataFrame({
    'Average Points': avg_points_by_variety.nlargest(5),
    'Frequency': data_subset['variety'].value_counts().loc[avg_points_by_variety.nlargest(5).index]
})
print(top_variety_df)
```
